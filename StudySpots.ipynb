```python
import pandas as pd
import numpy as np

df = pd.read_csv('HN Study Spots (Responses) - Form Responses 1.csv')
df['What is your favorite study spot on campus?'] = np.where((df['What is your favorite study spot on campus?'].str.lower() != 'snell library')
& (df['What is your favorite study spot on campus?'].str.lower() != 'exp') & (df['What is your favorite study spot on campus?'].str.lower() != 'isec'),
                                                             'other', df['What is your favorite study spot on campus?'])
counts_series = df['What is your favorite study spot on campus?'].str.lower().value_counts()
counts_series
counts_df = counts_series.reset_index()
counts_df.columns = ['Study Spot', 'Count']
counts_df
counts_df.to_csv('study_spot_counts.csv', index=False)
```
